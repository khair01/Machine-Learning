```python
ModelData = pd.read_csv('/content/drive/My Drive/Salary_Data.csv')
```

Splitting Data in x and y

```python
# Splitting data in Variables
x = ModelData.iloc[:,  :1]  # independent
y = ModelData.iloc[:, 1:]  # dependent
```

```python
plt.scatter(x, y, color = 'red')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
# Regression model
regression = LinearRegression()
regression.fit(x, y)
y_pred = regression.predict(x)
```

```python
plt.scatter(x, y, color = "red")
plt.plot(x,y_pred,color = "black")
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
